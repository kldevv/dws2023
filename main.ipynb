# Map

```python
title = 'Billboards Scatter Plot in New York City'
attr = f'<h6><b>{title}</b></h6> Map tiles by <a href="https://stamen.com/">Stamen Design</a>, under <a href="https://creativecommons.org/licenses/by/3.0">CC BY 3.0</a>. Data by <a href="https://data.cityofnewyork.us/">NYC OpenData</a>'
tiles = r'https://stamen-tiles-{s}.a.ssl.fastly.net/toner-lite/{z}/{x}/{y}.png'
nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=12, tiles=tiles, attr=attr)

# Add billboards makers
def add_marker(row):
    folium.Marker(location=[row['Lat.'], row['Long.']], icon=folium.Icon(color="red", icon="info-sign")).add_to(nyc_map)
billboards.apply(add_marker, axis=1)

# Display
nyc_map
```
